```python
#This dataset comprises of various house listings in London and neighbouring region. It also have the parameters listed below,
#the definitions of which are quite self-explanatory.
#• Property Name
#• Price
#• House Type - Contains one of the following types of houses (House, Flat/Apartment, New Development, Duplex, Penthouse, Studio, Bungalow, Mews)
#• Area in sq ft
#• No. of Bedrooms
#• No. of Bathrooms
#• No. of Receptions
#• Location
#• City/County - Includes London, Essex, Middlesex, Hertfordshire, Kent, and Surrey.
#• Postal Code
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("London.csv")
df = pd.read_csv("C:/Users/Effie/Downloads/London.csv")
df
```

```python
index_names = df[df['City/County'] != 'London'].index #delete the rows with location is not in london 
  
# drop these given row
# indexes from dataFrame
df.drop(index_names, inplace = True)
  
df
```

```python
duplicatedvalue = df[df.duplicated()] #check duplicated values
duplicatedvalue
```

```python
duplicatedvalue = df[df[['Property Name','Price']].duplicated()] 
duplicatedvalue
#We do have duplicates but the property name for each row is different. 
#This means the same location/price can have different houses, thus we cannot remove these duplicates
```

```python
#add a column to calculate the area in sq meter
df['Area in sq meter'] = round(df['Area in sq ft']*0.092903,1)#round to 1 decimal place
df
```

```python
#add a column to calculate the price per sq meter
df['Price per sq meter'] = round(df['Price']%df['Area in sq meter'],1)
df
```

```python
#Top 10 locations with lowest housing price (on average)
df=df.groupby('Location')['Price'].agg(['mean']).sort_values(by=('mean'),ascending=True).reset_index().head(10)

fig1, ax1 = plt.subplots(figsize=(18, 6))
ax1.set_facecolor('dimgrey')
ax1.bar(df['Location'],df['mean'],color='lightyellow')
plt.title('Top 10 locations with lowest house price')
plt.xlabel("Location")
plt.ylabel("Price")
plt.show()
```

```python
#Top locations with lowest housing price (on average)
df=df.groupby('Location')['Price'].agg(['mean']).sort_values(by=('mean'),ascending=False).reset_index().head(10)

fig1, ax1 = plt.subplots(figsize=(18, 6))
ax1.set_facecolor('dimgrey')
ax1.bar(df['Location'],df['mean'],color='lightyellow')
plt.title('Top locations with lowest house price')
plt.xlabel("Location")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.scatter(df['Area in sq ft'],df['Price'], label = "Area in sq ft")
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("The Relationship Between house area and price")
plt.show()
```

```python
#create a grouped bar chart for the top 10 locations with lowest house price
#bar1:average house price
#bar2:average price per sq meter
#bar3:average area in sq meter




#create the base axis
fig, ax = plt.subplots(1,1, figsize = (8,6))
#set the labels
#and the x positions
label = df["Postal Code"]
x = np.arange(len(label))
#set the width of the bars
width = 0.2
#create the first bar -1 width
rect1 = ax.bar(x - width,
              #df[2012],
        df=df.groupby('Postal Code')['Price'].agg(['mean']).sort_values(by=('mean'),ascending=False).reset_index().head(10),
              width = width, 
               label = "average house price",
               edgecolor = "black")




#create the second bar using x
rect2 = ax.bar(x,
              #data_df[2013],
df=df.groupby('Postal Code')['Price per sq meter'].agg(['mean']).sort_values(by=('mean'),ascending=False).reset_index().head(10),
              width = width,
              #label = 2013,
               label="average price per sq meter",
              edgecolor = "black")


#create the third bar plus 1 width
rects2 = ax.bar(x + width,
              # data_df[2014],
df=df.groupby('Postal Code')['Area in sq meter'].agg(['mean']).sort_values(by=('mean'),ascending=False).reset_index().head(10), 
               width = width,
               label = "average area in sq meter",
               edgecolor = "black")

#add the labels to the axis
ax.set_ylabel("Price in £,Area in cm2",
             fontsize = 25,
             labelpad = 20)
ax.set_xlabel("Post Code",
             fontsize = 25,
             labelpad =20)
ax.set_title("Top 10 cheapest area to buy a property in London",
            fontsize = 30,
            pad = 20)
#set the ticks
#using the labels
ax.set_xticks(x)
ax.set_xticklabels(label)
#add the legend
ax.legend(#title = "Year of election",
         fontsize = 16,
         title_fontsize = 20,
         bbox_to_anchor = (1.02, 0.7))
#adjust the tick paramaters
ax.tick_params(axis = "x",
              which = "both",
              labelrotation = 90,
              labelsize = 20)
ax.tick_params(axis = "y",
              which = "both",
              labelsize = 20)
```

```python
# Import Library

import numpy as np 
import matplotlib.pyplot as plt 

#create a grouped bar chart for the top 10 locations with lowest house price
#bar1:average house price
#bar2:average price per sq meter
#bar3:average area in sq meter

# Define Data


location=df.groupby('Postal Code')['Price'].agg(['mean']).sort_values(by=('mean'),ascending=False).reset_index().head(10)
averageprice=df.groupby('Postal Code')['Price'].agg(['mean']).sort_values(by=('mean'),ascending=False).reset_index().head(10)
pricepersq=df.groupby('Postal Code')['Price per sq meter'].agg(['mean']).sort_values(by=('mean'),ascending=False).reset_index().head(10)  
area=df.groupby('Postal Code')['Area in sq meter'].agg(['mean']).sort_values(by=('mean'),ascending=False).reset_index().head(10)


x_axis = np.arange(len(location))

# Multi bar Chart

plt.bar(x_axis +0.20, averageprice, width=0.2, label = 'average house price')
plt.bar(x_axis +0.20*2, pricepersq, width=0.2, label = 'average price per sq meter')
plt.bar(x_axis +0.20*3, area, width=0.2, label = 'average area in sq meter')

# Xticks

plt.xticks(x_axis,location)

# Add legend

plt.legend()

# Display

plt.show()
```
